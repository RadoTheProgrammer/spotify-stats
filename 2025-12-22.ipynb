```python
import pandas as pd
n = 100
df = pd.read_csv("StreamingHistory.csv", parse_dates=["endTime"])
```

```python
grouped = df.groupby(df.index // n)
```

```python
first_endtime = grouped["endTime"].first()
```

```python
first_endtime
```

```python
value_counts = grouped.value_counts(["artistName"])
value_counts = grouped.value_counts(["artistName","trackName"])
```

```python
value_counts_group = value_counts.groupby(level=0)

value_counts_head = value_counts_group.head(5)
value_counts_head = value_counts_head.reset_index()
#value_counts_head["endTime"] = ""
value_counts_head.loc[::5,"endTime"] = first_endtime.values
value_counts_head
```

```python
value_counts_head.to_csv("group_counts_track.csv", index=False)
```

```python
import pandas as pd

# original dataframe
df = pd.DataFrame({'col1': [1, 2, 3, 4]})

# series you want to insert every 2 rows
s = ["A", "B"]

# create empty column
df["col2"] = ""

# assign values every 2 rows
df.loc[::2, "col2"] = s

print(df)
```
